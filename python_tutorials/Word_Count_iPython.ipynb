```python
#Import Libraries

import operator as op

#Function "gen_Word_Count" will break down larg file into words one by one 

def gen_Word_Count(path):
    word_list = []
    read_file = open(path, 'r') # Read File
    for word in read_file:
        content = word.lower().split()  # Make sure all the words are consistant and split by space
        for each_word in content:
            word_list.append(each_word) # Adds words to our List
        #print(word_list)
    remove_puncatuation(word_list)

#Function "remove_puncatuation" will remove punctuation from the list of words

def remove_puncatuation(word_list):
    clean_word_list = []
    for word in word_list:
        symbols = ",;:.'`"  # List of minimum punctuation Marks
        for i in range(0,len(symbols)):  # Check for every word if there is any punctuation 
            word = word.replace(symbols[i], "")  # Replace it with nothing
        if len(word) >0:         # Check there may be a special Character which actually not a word
            clean_word_list.append(word)
    #print(clean_word_list)
    creat_dictionary(clean_word_list)

#Function "creat_dictionary" will count the words occurance and create dictionary of the words

def creat_dictionary(clean_word_list):
    word_count = {}
    #print(clean_word_list)
    for word in clean_word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    for key, value in sorted(word_count.items(), key=op.itemgetter(1)): # Sort words according to the Count
        print(key," : ", value)

file_path ="E:\Workspace\Learning\ML_Lab_1\ML_data.txt"    # change the path according to desired path
gen_Word_Count(file_path) # Call the function 
```
